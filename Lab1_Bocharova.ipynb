```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def task_1():
    even_number = int(input("Введіть парне число (окрім 2): "))
    
    if even_number % 2 != 0 or even_number == 2:
        print("Будь ласка, введіть парне число (окрім 2).")
        return

    for i in range(2, even_number):
        if is_prime(i) and is_prime(even_number - i):
            print(f"Число {even_number} можна представити як {i} + {even_number - i}")
            return

    print(f"Не вдалося знайти два простих числа, які утворюють суму {even_number}.")

task_1()
```
