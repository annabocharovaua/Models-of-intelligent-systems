```python
# Transformation Pipelines

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

data_num_tr = num_pipeline.fit_transform(data_num)

#Цей код використовує трансформаційний конвеєр (pipeline) для обробки числових даних. 
#Перш за все, використання конвеєру спрощує та систематизує процес обробки даних. 
#Замість того, щоб виконувати кожен етап обробки окремо, такі як заповнення пропущених значень або стандартизація даних, 
#можна об'єднати їх у логічні послідовності за допомогою конвеєру. Це полегшує розуміння та підтримку коду.
#По-друге, застосування конвеєру дозволяє уникнути витоку інформації про тестові дані під час навчання моделі.
#Тобто параметри трансформації (наприклад, середнє значення для заповнення пропущених даних) обчислюються лише на основі
#навчальних даних, а потім використовуються для трансформації як навчальних, так і тестових даних. 
#Це дозволяє уникнути перенавчання та забезпечує більш об'єктивні результати.
#Отже, використання трансформаційного конвеєру у коді допомагає створити структурований та ефективний підхід
#до обробки даних у машинному навчанні, забезпечуючи при цьому правильне розподілення і обробку інформації 
#для побудови моделей з високою точністю.
```

```python
# ColumnTransformer
from sklearn.compose import ColumnTransformer

num_attribs = list(data_num)
cat_attribs = ["date"]
full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(handle_unknown='ignore'), cat_attribs),
    ])

data_prepared = full_pipeline.fit_transform(data)
#Використання ColumnTransformer дозволяє систематизувати та автоматизувати процес обробки різних типів атрибутів
#у наборі даних перед їх використанням для навчання моделей машинного навчання. 
#Цей підхід дозволяє одночасно застосовувати різні методи обробки (такі як заповнення пропущених значень,
#шкалювання, кодування категоріальних змінних тощо) до відповідних стовпців даних, забезпечуючи при цьому правильну обробку 
#кожного типу атрибута. Такий підхід допомагає уникнути витоку інформації про тестові дані та забезпечує стабільні
#та об'єктивні результати моделювання в машинному навчанні.
```

```python
# Scikit-Learn’s mean_squared_error()

from sklearn.metrics import mean_squared_error

data_predictions = lin_reg.predict(data_prepared)
lin_mse = mean_squared_error(data_labels, data_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
# Результатом є значення MSE, яке представляє собою середню квадратичну відстань між прогнозованими та фактичними значеннями. 
#Далі, використовуючи це значення, ми обчислюємо квадратний корінь з MSE (RMSE), що є показником помилки моделі:
#чим менше значення RMSE, тим краще модель відповідає даним.
```

# DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(data_prepared, data_labels)

# evaluate the model on training set
data_predictions = tree_reg.predict(data_prepared)
tree_mse = mean_squared_error(data_labels, data_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse

# У цьому фрагменті коду ми використовуємо модель регресії на основі дерева прийняття рішень для навчання наших даних
#та оцінки її ефективності. Спочатку ми створюємо екземпляр моделі DecisionTreeRegressor(),
#потім застосовуємо метод fit() для навчання моделі на вхідних даних,підготовлених за допомогою конвеєру підготовки даних.
#Потім ми використовуємо навчену модель для прогнозування цін на нерухомість для того ж самого набору даних,
#на якому вона була навчена. Оцінюємо ефективність моделі, обчислюючи середньоквадратичну помилку (RMSE) між прогнозованими 
#і фактичними значеннями.В результаті отримуємо значення RMSE 0. модель показує сильне перенавчання на навчальних даних,
#оскільки RMSE на навчальному наборі даних є 0, що свідчить про високу точність моделі на цих даних,
#але модель може показувати погані результати на нових, тестових даних.
```

```python
# Better Evaluation Using Cross-Validation
# train_test_split() OR Scikit-Learn’s K-fold cross-validation

from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, data_prepared, data_labels,
scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
display_scores(tree_rmse_scores)
#У цьому фрагменті коду ми використовуємо кращий метод оцінки ефективності моделі за допомогою перехресної перевірки.
#Спочатку ми використовуємо функцію cross_val_score() з бібліотеки Scikit-Learn для оцінки ефективності моделі регресії
#на основі дерева прийняття рішень за допомогою перехресної перевірки. Ми вказуємо модель, дані та мітки, 
#а також кількість згорток (у нашому випадку 10) за допомогою параметру cv.
#Потім ми обчислюємо квадратний корінь з середньоквадратичної помилки (RMSE) для кожної згортки та виводимо результати, 
#включаючи середнє та стандартне відхилення оцінок RMSE для всіх згорток.
#Цей підхід дозволяє отримати більш надійну оцінку ефективності моделі, оскільки вона оцінюється на декількох незалежних наборах даних.
```

```python
# Cross-Validation for Linear Regression

lin_scores = cross_val_score(lin_reg, data_prepared, data_labels,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
#У цьому фрагменті коду ми використовуємо перехресну перевірку для оцінки ефективності моделі лінійної регресії. 
#Використовуючи функцію cross_val_score() з бібліотеки Scikit-Learn, ми оцінюємо модель на кожній згортці та обчислюємо
#середньоквадратичну помилку (RMSE) для кожної згортки. 
```

# Fine-Tune The Model

## Grid Search

```python
# For RandomForestRegressor

from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                            scoring='neg_mean_squared_error',
                            return_train_score=True)

grid_search.fit(data_prepared, data_labels)
#У даному фрагменті коду ми використовуємо Grid Search для пошуку найкращих гіперпараметрів моделі RandomForestRegressor. 
#Ми визначаємо набір можливих значень для параметрів моделі, таких як кількість дерев (n_estimators)
#та максимальна кількість ознак (max_features). Параметри обертаються по заданих значеннях, а результати оцінюються
#за допомогою перехресної перевірки з використанням середньоквадратичної помилки як метрики оцінки.
#Після завершення пошуку найкращих параметрів за допомогою Grid Search, ми отримуємо навчену модель, 
#яка оптимізована для найкращих параметрів.
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
#У цьому фрагменті коду ми виводимо результати крос-валідації для кожного набору гіперпараметрів,
#які були перевірені за допомогою Grid Search. Для кожного набору гіперпараметрів ми виводимо середнє значення RMSE
#(квадратний корінь з від'ємного середньоквадратичного значення помилки) з результатів перехресної перевірки,
#а також самі гіперпараметри, що використовувалися. Це дозволяє нам порівняти ефективність моделі з різними наборами параметрів
#та обрати оптимальний набір для подальшого використання.
# можна побачити, що комбінація гіперпараметрів max_features=8 та n_estimators=30 має найменше значення RMSE,
#що свідчить про кращу точність моделі на навчальних даних. 
```

```python
# For RandomForestRegressor

from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    "max_depth": [3, 5, 10, None],
    "n_estimators": [30, 100, 200, 300, 400, 500],
    "max_features": [np.random.randint(1, 30) for _ in range(30)],
    "bootstrap": [True, False],
    "min_samples_leaf": [np.random.randint(1, 40) for _ in range(40)]
}

forest_reg = RandomForestRegressor()

random_search = RandomizedSearchCV(
    forest_reg, 
    param_distributions=param_dist, 
    n_iter=100,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=-1,  # Use all available CPU cores for parallelization
    return_train_score=True
)

random_search.fit(data_prepared, data_labels)
#У даному фрагменті коду ми використовуємо випадковий пошук для знаходження найкращих гіперпараметрів моделі
#RandomForestRegressor. Ми визначаємо простір пошуку, включаючи діапазон значень для кожного гіперпараметра,
#такі як max_depth, n_estimators, max_features, bootstrap та min_samples_leaf. Після визначення простору пошуку,
#ми використовуємо функцію RandomizedSearchCV() для випадкового пошуку по цьому просторі з 100 ітераціями та крос-валідацією 
#на 5 складових. Ми оцінюємо кожну комбінацію гіперпараметрів за допомогою середньоквадратичної помилки та шукаємо комбінацію,
#яка має найкращий результат.
```
